```python
import numpy as np
import matplotlib.pyplot as plt
```

1 , 2) Generate a random 2d matrix having 200 rows

```python
np.random.seed(10)
x = np.random.rand(200, 1)
```

3) Formulate a linear regression line (model) having hypothesis: y=10+7

     [ Linear Regression finds a straight line that best fits the relationship between a predicted variable (y_predicted) and a continuous target variable (y) ]

```python
y=10*x+7
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

4) Find mean & standard deviation of row elements

```python
mean = x.mean()
std = x.std()
print("Mean:", mean)
print("Standard Deviation:", std)
```

5) Fit linear regression model and train with row elements

```python
model.fit(x,y)
```

6) Predict "y-predicted" values using regression model

```python
y_predicted = model.predict(x)
```

7) Evaluate rms (root mean square) error between actual level i.e. 'y' and predicted level 'y-predicted'

```python
from sklearn.metrics import mean_squared_error as mse
rmse = np.sqrt(mse(y,y_predicted))
rmse
```

8) Find out intercept of hypothesis

```python
print("Intercept of the hypothesis :", model.intercept_[0])
```

9) find out slope of hypothesis

```python
print("Slope of the hypothesis :", model.coef_[0][0])
```

10) Draw curve of hypothesis

```python
plt.plot(y,y_predicted,color='g')
plt.xlabel('y')
plt.ylabel('y_predicted')
plt.title('Hypothesis Plot')
plt.show()
```
