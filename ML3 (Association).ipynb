b) Generate list of transactions from the dataset

```python
transactions = []
for i in range(len(data)):
    transactions.append([str(data.values[i, j]) for j in range(len(data.columns)) if str(data.values[i, j]) != 'nan'])
```

c) Train Apriori algorithm on the dataset   (Find frequent itemsets using apriori

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

one_hot_encoded = pd.get_dummies(pd.DataFrame(transactions), prefix='', prefix_sep='')
frequent_itemsets = apriori(one_hot_encoded, min_support=0.005, use_colnames=True)
```

d) Visualize the list of rules

```python
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
print(rules)
```

e) Find new rules by increasing minimum confidenc

```python
new_rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.3)
print(new_rules)
```
