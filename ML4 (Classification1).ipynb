a.2) create new column 'spam' in dataframe df of type 'int' where 0 means false (no spam) and 1 means true (spam)

```python
import numpy as np
df['spam']=(df['Category']=='spam').astype(np.int8)
```

a.3) identify columns with missing values

```python
df.columns[df.isna().any()]
```

b) Data preparation (split data for training and testing)

```python
from sklearn.model_selection import train_test_split as tts
X_train,X_test,y_train,y_test = tts(df.Message,df.spam,test_size=0.2,random_state=42)
```

c) Apply Multinomial Naive bayes algorithm to classify SPAM message with CountVectorizer for feature Extraction

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
clf=Pipeline([
    ('vectorizer',CountVectorizer()),
    ('nb',MultinomialNB())
])
```

** fit ML pipeline on training data

```python
clf.fit(X_train,y_train)
```

d) Evaluate model

```python
emails=[
    'hey mohan,can we meet?',
    'upto 20% discount on parking, dont miss the order'
]
clf.predict(emails)
```

e) Compare performance

```python
clf.score(X_test,y_test)
```
